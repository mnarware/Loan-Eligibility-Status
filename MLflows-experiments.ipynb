```python
import numpy as np
import pandas as pd
```

```python
train_df=pd.read_csv("data.csv")
train_df.head()
```

```python
train_df.shape
```

```python
train_df["Gender"]=train_df["Gender"].map({"Male":0,"Female":1})
train_df["Married"]=train_df["Married"].map({"Yes":1,"No":0})
train_df["Loan_Status"]=train_df["Loan_Status"].map({"Y":1,"N":0})
train_df.head()
```

```python
train_df.info()
```

```python
train_df.isnull().sum()
```

```python
train_df=train_df.dropna()
train_df.isnull().sum()
```

```python
train_df.shape
```

```python
X=train_df[['Gender','Married','ApplicantIncome','LoanAmount','Credit_History']]
y=train_df.Loan_Status
X.shape,y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(max_depth=4,random_state=5)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
pred_val=model.predict(X_val)
accuracy_score(y_val,pred_val)
```

```python
pred_train=model.predict(X_train)
accuracy_score(y_train,pred_train)
```

```python
#Run with function with parameter
```

```python
def mlflow_runs(n_est,max_deep,i):
    with mlflow.start_run():
        model=RandomForestClassifier(max_depth=max_deep,n_estimators=n_est, random_state=5)
        model.fit(X_train,y_train)  #New Data
        pred_val=model.predict(X_val)
        val_acc=accuracy_score(y_val,pred_val)
    
        pred_train=model.predict(X_train)
        train_acc=accuracy_score(y_train,pred_train)
    
        #tag Name column name and column data dislpay info
        runs="hyperparameter"+str(i)
        mlflow.set_tag('mlflow.runName',runs)
        mlflow.set_tag('Algorithm Name',"RandomForestClassifier")
        mlflow.set_tag("Data","Old Data")
        
        #log the hyperparameter tuning
        mlflow.log_param("max_depth",max_deep)
        mlflow.set_tag("n_estimators",n_est) 
        
        # log performance metrics
        mlflow.log_metric("train_accuracy",train_acc)
        mlflow.log_metric("val_accuracy",val_acc)
        
        #log teh entire model
        mlflow.sklearn.log_model(model,"myModel")
```

```python
n_estimators=[10,20,30]
max_depths=[2,4,6,8]
count=0
for i in n_estimators:
    for j in max_depths:
        count+=1
        mlflow_runs(i,j,count)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
with mlflow.start_run():
    model=KNeighborsClassifier(n_neighbors=5)
    model.fit(X_train,y_train)  #New Data
    pred_val=model.predict(X_val)
    val_acc=accuracy_score(y_val,pred_val)

    pred_train=model.predict(X_train)
    train_acc=accuracy_score(y_train,pred_train)

    #tag Name column name and column data dislpay info
    mlflow.set_tag('mlflow.runName',"KNC Run")
    mlflow.set_tag('Algorithm Name',"KNeighborsClassifier")
    mlflow.set_tag("Data","New data")
    
    
    
    #log the hyperparameter tuning
    mlflow.log_param("n_neighbors",5)
    mlflow.log_metric("train_accuracy",train_acc)
    mlflow.log_metric("val_accuracy",val_acc)
    

    cm=ConfusionMatrixDisplay.from_predictions(y_val,pred_val)
    cm.figure_.savefig("./confusion_matrix")

    #log teh entire model
    mlflow.sklearn.log_model(model,"KNC Model")
```
